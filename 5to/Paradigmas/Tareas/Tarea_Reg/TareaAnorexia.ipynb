```r
library(MASS)
library(ggplot2)

ggplot(anorexia, aes(x = Prewt, y = Postwt, shape = Treat, color = Treat)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE) +
  facet_grid(. ~ Treat) +
  theme_classic()
```
