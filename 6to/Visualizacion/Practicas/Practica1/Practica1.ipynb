# Ejercicio 2 

A partir del archivo empleo.csv realizar una visualización donde se compraren los diferentes
tipos de ingreso entre hombres y mujeres para los años 2020, 2021, 2022, y 2023. Se sugiere una
grafica de barras agrupadas [1] o una gráfica de barras en espejo [2].

```python
import pandas as pd 

emp = pd.read_csv('./empleo - empleo.csv')
emp.head()
```

```python
#Grupamos cada año y su valor
emp['a_2020'] = emp['2020/t1'] + emp['2020/t3'] + emp['2020/t4']
emp['a_2021'] = emp['2021/t1'] + emp['2021/t2'] + emp['2021/t3'] + emp['2021/t4']
emp['a_2022'] = emp['2022/t1'] + emp['2022/t2'] + emp['2022/t3'] + emp['2022/t4'] 
emp['a_2023'] = emp['2023/t1'] + emp['2023/t2'] + emp['2023/t3'] 
```

```python
años = emp['a_2020'] + emp['a_2021'] + emp['a_2022'] + emp['a_2023']
```

```python
#Filtramos por sexo y agregamos una columna que sea el sexo
sexo = []
for index, row in emp.iterrows():
    ind = row['indicador']
    if 'hombres' in ind:
        sexo.append('hombre')
    elif 'mujeres' in ind:
        sexo.append('mujer')
    else:
        sexo.append(None)

emp['sexo'] = sexo
emp = emp[emp['sexo'].notnull()] # Eliminaos , los que no tienen valor por tener el nombre de ambos
#Solo nos quedamos con los años y su poblacion total del año 
emp_años = emp.drop(columns=['2020/t1', '2020/t3', '2020/t4','2021/t1', '2021/t2', '2021/t3', '2021/t4','2022/t1', '2022/t2', '2022/t3', '2022/t4','2023/t1', '2023/t2', '2023/t3'])
#Quitamos la frafe de mas de 15 para que no se vea mucho rexto en la grafica
emp_años['indicador'] = emp_años['indicador'].str.replace(' - 15 años y más, hombres', '', regex=False)
emp_años['indicador'] = emp_años['indicador'].str.replace(' - 15 años y más, mujeres', '', regex=False)
emp_años.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(50, 25))

# Crear gráfico de barras
sns.barplot(data=emp_años, x="indicador", y="a_2020", hue="sexo", palette="Set1")
plt.title("Población ocupada por sexo en 2020", fontsize=30)
plt.xlabel("Grupo de ingresos", fontsize=40)
plt.ylabel("Población", fontsize=12)
plt.xticks(rotation=20, fontsize = 20)
plt.legend(title="Sexo", fontsize = 40)

# Mostrar
plt.tight_layout()
plt.show()
```

# Ejercicio 3

A partir de la tabla conjunto_de_datos_defunciones_registradas_2022.csv generar una visualización de un mapa con burbujas que permita visualizar las muertes de mujeres causadas por violencia familiar en México para el año 2022, para los datos totales y por cada 100,000 habitantes.

```python
import pandas as pd 

municipios = pd.read_csv("./Municipios.csv", encoding="latin1") # Municipio de cada estado con latitud y longitud (esto lo ocupe el smestre pasado y lo obtuve de pagina de gobierno arriba el link de donde extraje los datos)
defun = pd.read_csv("./conjunto_de_datos_defunciones_registradas_2022.CSV")
```

```python
municipios.rename(columns={"ides": "ent_regis", "idmun": "mun_regis"}, inplace=True)
defun = defun.merge(municipios[['ent_regis', 'mun_regis', 'nmun', 'lat', 'lon']], 
                     on=['ent_regis', 'mun_regis'], how='left')

defun.head() #Agregamos la latitud y longitud de cada municipio
```

```python
c_def = defun[defun['causa_def'].between("X850", "Y099")] # Filtardo  por causa de defunción

#condición de violencia familiar en las defuncionescondición de violencia familiar en las defunciones
p_agre_existe = c_def[c_def['par_agre'].between(1, 26) | c_def['par_agre'].between(37, 52)] # (existio violencia)
p_agre_existe_muj = p_agre_existe[p_agre_existe['sexo'] == 2] 
```

```python
#Datos obtenidos del INEGI del censo del 2020
Poblacion = {'Estado': 'Poblacion' ,'1': 1425607, '2': 3769020, '3': 798447, '4': 928363, '5':3146771,
    '6': 731391, '7': 5543828, '8': 3741869, '9': 9209944, '10': 1832650, '11':6166934,
    '12': 3540685, '13': 3082841, '14': 8348151, '15': 16992418, '16': 4748846, '17': 1971520,
    '18': 1235456, '19': 5784442, '20': 4132148, '21': 6583278, '22': 2368467, '23': 1857985,
    '24': 2822255, '25': 3026943, '26': 2944840, '27': 2402598, '28': 3527735, '29': 1342977,
    '30': 8062579, '31': 2320898	, '32':1622138
}
df_poblacion = pd.DataFrame(list(Poblacion.items()), columns=['Estado', 'Poblacion'])
df_poblacion = df_poblacion.drop(0)
```

```python
mapa_fin = p_agre_existe_muj.groupby('ent_regis').size().reset_index()
mapa_fin.columns = ["Estado", "Muertes"] 
mapa_fin['Estado'] = mapa_fin['Estado'].astype(str)
df_poblacion['Estado'] = df_poblacion['Estado'].astype(str)
mapa_fin_1 = mapa_fin.merge(df_poblacion, on='Estado', how='left')
mapa_fin_1['Tasa'] = (mapa_fin_1['Muertes'] / mapa_fin_1['Poblacion']) * 100000
mapa_fin_1.head()
```

```python
defun_lat = defun.groupby('ent_regis').first().reset_index()

mapa_fin_1['Estado'] = mapa_fin_1['Estado'].astype(str)
defun_lat['ent_regis'] = defun_lat['ent_regis'].astype(str)

mapa_fin_1 = mapa_fin_1.merge(defun_lat[['ent_regis', 'lat', 'lon']], 
                               left_on='Estado', right_on='ent_regis', 
                               how='left')
mapa_fin_1.head()
```

```python
import folium
import math

m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
# add marker one by one on the map, and account for Mercator deformation
for _, estado in mapa_fin_1.iterrows():
    local_deformation = math.cos(estado['lat'] * math.pi / 180)
    folium.Circle(
        location=[estado['lat'], estado['lon']],
        popup='%s (%.1f)' % (estado['Estado'], estado['Tasa']),
        radius=estado['Muertes'] * 25000 * local_deformation,
        color='crimson',
        fill=True,
        fill_color='crimson'
    ).add_to(m)

# Show the map again
m
```
