### Inciso 3.1

Considera una v.a. $X \sim \mathcal{U}(0,10)$, calcula su esperanza $(\mu)$ y varianza $\left(\sigma^{2}\right)$. Fija $n=5$ y realiza los siguientes pasos:

- Esperanza $(\mu)$: Sabemos que $X \sim \mathcal{U}(0, 10)$ por lo que podemos obtenerla mediante
  $\mu = \frac{a + b}{2}$

  Ahora bien como tenemos que a = 0 y b = 10 , sustituyendo tenemos que : 

  $\mu = \frac{0 + 10}{2} = 5$

- Varianza $\left(\sigma^{2}\right)$: Al tratarse de una uniforme sabemos    que 

  $\sigma^{2} = \frac{(b - a)^2}{12}$

  Por lo que sustituyendo los valores tenemos que 

  $\sigma^{2} = \frac{(10 - 0)^2}{12} \approx 8.333$


1) Genera una m.a. $\left(X_{1}, X_{2}, \ldots, X_{n}\right)=\left(x_{1}^{1}, x_{2}^{1}, \ldots, x_{n}^{1}\right)$ de $X$.

```python
import numpy as np 

a = 0 
b = 10
n = 5 # Tamaño de la muestra alatoria 

muestra = np.random.uniform(a,b,n)

print('La muestra aletoria generada es :', muestra)
```

2) Calcula la media observada $\bar{x}_{n}^{1}$.

```python
media_obs = np.mean(muestra)

print('La media observada es:', media_obs)
```

3) Repite los dos pasos anteriores $m=10,000$ veces para obtener $\bar{x}_{n}^{1}, \bar{x}_{n}^{2}, \ldots, \bar{x}_{n}^{m}$.

```python
m = 10000 # Veces a repetir el proceso

medias_obs = []

for i in range(m):
    muestra = np.random.uniform(a,b,n)
    media_obs = round(np.mean(muestra), 3)
    medias_obs.append(media_obs)

medias_obs = np.array(medias_obs)

print('Las primeras 15 medias observadas son:', medias_obs[:10])
```

4) Haz un histograma de las medias observadas $\bar{x}_{n}^{1}, \bar{x}_{n}^{2}, \ldots, \bar{x}_{n}^{m}$.

```python
import matplotlib.pyplot as plt
plt.hist(medias_obs, bins= 30, density = True, edgecolor = 'black', alpha = 0.7)
plt.title("Histograma de las medias observadas")
plt.xlabel("Media observada")
plt.ylabel("Densidad")
plt.show()
```

5) Encima del histograma, grafica la función de densidad de probabilidad de una $\mathcal{N}\left(\mu, \frac{\sigma^{2}}{n}\right)$. En estadística, coloquialmente se dice que estamos "ajustando" la distribución normal a los datos. 

```python
from scipy.stats import norm
mu = 5  
sigma_cuad = (b - a)**2 / 12  
sigma_cuad_media = sigma_cuad / n
sigma_media = np.sqrt(sigma_cuad_media)  

# Crear el histograma
plt.hist(medias_obs, bins=30, density=True, edgecolor='black', alpha=0.7, label="Medias observadas")

# Crear un rango de valores para la PDF teórica
x = np.linspace(mu - 4*sigma_media, mu + 4*sigma_media, 1000)
pdf_teorica = norm.pdf(x, mu, sigma_media)
plt.plot(x, pdf_teorica, 'r-', label="PDF teórica (Normal)")

# Añadir título, etiquetas y leyenda
plt.title("Histograma de las medias observadas y ajuste normal")
plt.xlabel("Media observada (x̄ₙ)")
plt.ylabel("Densidad")
plt.legend()

# Mostrar el gráfico
plt.show()
```

6) Repite toda la simulación para $n=30,100,200$ y presenta las cuatro gráficas en una matriz de $2 \times 2$ para poder apreciar el impacto de aumentar el tamaño de muestra.

```python
muestras = [5, 30, 100, 200] 

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Distribución de las medias observadas para diferentes tamaños de muestra", fontsize=16)

# Función para generar medias observadas y graficar
def graficar_histograma_y_pdf(ax, n):
    # Lista para almacenar las medias observadas
    medias_observadas = []

    # Repetir el proceso m veces
    for i in range(m):
        # Generar una muestra aleatoria de tamaño n
        muestra = np.random.uniform(a, b, n)
        
        # Calcular la media observada
        media_observada = np.mean(muestra)
        
        # Guardar la media observada en la lista
        medias_observadas.append(media_observada)

    # Convertir la lista a un array de numpy
    medias_observadas = np.array(medias_observadas)

    # Parámetros de la distribución normal teórica
    mu = 5  # Esperanza teórica
    sigma_cuad = (b - a)**2 / 12  # Varianza de la distribución uniforme
    sigma_cuad_media = sigma_cuad / n  # Varianza de la media muestral
    sigma_media = np.sqrt(sigma_cuad_media)  # Desviación estándar de la media muestral

    # Crear el histograma
    ax.hist(medias_observadas, bins=30, density=True, edgecolor='black', alpha=0.7, label="Medias observadas")

    # Crear un rango de valores para la PDF teórica
    x = np.linspace(mu - 4*sigma_media, mu + 4*sigma_media, 1000)
    pdf_teorica = norm.pdf(x, mu, sigma_media)

    # Superponer la PDF teórica sobre el histograma
    ax.plot(x, pdf_teorica, 'r-', label="PDF teórica (Normal)")

    # Añadir título y etiquetas
    ax.set_title(f"n = {n}")
    ax.set_xlabel("Media observada (x̄ₙ)")
    ax.set_ylabel("Densidad")
    ax.legend()

# Graficar para cada tamaño de muestra
for i, n in enumerate(muestras):
    row = i // 2  # Fila en la matriz de subgráficos
    col = i % 2   # Columna en la matriz de subgráficos
    graficar_histograma_y_pdf(axs[row, col], n)

plt.tight_layout()
plt.show()
```

### Inciso 3.2

Considera una v.a. $X \sim \operatorname{Bernoulli}(\theta)$, calcula su esperanza $(\mu)$ y varianza $\left(\sigma^{2}\right)$. Fija $\theta=0.05, n=5$ y realiza los exactamente los mismos pasos que en el ejercicio 3.1.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros
theta = 0.05  # Parámetro de la distribución Bernoulli
n = 5         # Tamaño de la muestra
m = 10000     # Número de repeticiones

# Esperanza y varianza teóricas
mu = theta
sigma_cuad = theta * (1 - theta)

# Lista para almacenar las medias observadas
medias_observadas = []

# Repetir el proceso m veces
for i in range(m):
    # Generar una muestra aleatoria de tamaño n
    muestra = np.random.binomial(n=1, p=theta, size=n)
    
    # Calcular la media observada
    media_observada = np.mean(muestra)
    
    # Guardar la media observada en la lista
    medias_observadas.append(media_observada)

# Convertir la lista a un array de numpy
medias_observadas = np.array(medias_observadas)

# Parámetros de la distribución normal teórica
sigma_cuad_media = sigma_cuad / n  # Varianza de la media muestral
sigma_media = np.sqrt(sigma_cuad_media)  # Desviación estándar de la media muestral

# Crear el histograma
plt.hist(medias_observadas, bins=30, density=True, edgecolor='black', alpha=0.7, label="Medias observadas")

# Crear un rango de valores para la PDF teórica
x = np.linspace(mu - 4*sigma_media, mu + 4*sigma_media, 1000)
pdf_teorica = norm.pdf(x, mu, sigma_media)

# Superponer la PDF teórica sobre el histograma
plt.plot(x, pdf_teorica, 'r-', label="PDF teórica (Normal)")

# Añadir título, etiquetas y leyenda
plt.title(f"Distribución de las medias observadas (n = {n}, θ = {theta})")
plt.xlabel("Media observada (x̄ₙ)")
plt.ylabel("Densidad")
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
# Parámetros
theta = 0.05
m = 10000
tamanos_muestra = [5, 30, 100, 200]  # Tamaños de muestra a evaluar

# Crear una figura con subgráficos de 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Distribución de las medias observadas para diferentes tamaños de muestra (θ = 0.05)", fontsize=16)

# Función para generar medias observadas y graficar
def graficar_histograma_y_pdf(ax, n):
    # Lista para almacenar las medias observadas
    medias_observadas = []

    # Repetir el proceso m veces
    for i in range(m):
        # Generar una muestra aleatoria de tamaño n
        muestra = np.random.binomial(n=1, p=theta, size=n)
        
        # Calcular la media observada
        media_observada = np.mean(muestra)
        
        # Guardar la media observada en la lista
        medias_observadas.append(media_observada)

    # Convertir la lista a un array de numpy
    medias_observadas = np.array(medias_observadas)

    # Parámetros de la distribución normal teórica
    mu = theta
    sigma_cuad = theta * (1 - theta)
    sigma_cuad_media = sigma_cuad / n
    sigma_media = np.sqrt(sigma_cuad_media)

    # Crear el histograma
    ax.hist(medias_observadas, bins=20, density=True, edgecolor='black', alpha=0.7, label="Medias observadas")

    # Crear un rango de valores para la PDF teórica
    x = np.linspace(mu - 4*sigma_media, mu + 4*sigma_media, 1000)
    pdf_teorica = norm.pdf(x, mu, sigma_media)

    # Superponer la PDF teórica sobre el histograma
    ax.plot(x, pdf_teorica, 'r-', label="PDF teórica (Normal)")

    # Añadir título y etiquetas
    ax.set_title(f"n = {n}")
    ax.set_xlabel("Media observada (x̄ₙ)")
    ax.set_ylabel("Densidad")
    ax.legend()

# Graficar para cada tamaño de muestra
for i, n in enumerate(tamanos_muestra):
    row = i // 2  # Fila en la matriz de subgráficos
    col = i % 2   # Columna en la matriz de subgráficos
    graficar_histograma_y_pdf(axs[row, col], n)

# Ajustar el espaciado entre subgráficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```
