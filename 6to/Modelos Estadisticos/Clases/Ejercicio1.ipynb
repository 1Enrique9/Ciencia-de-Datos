```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(25)
y=np.random.randint (10, 20, 25)
y[24]=y[0]
fig = plt.figure(figsize=(7,5))
ax = plt.subplot(111)
ax.fill_between (x,y)
ax.set_xlabel('Hora de posteo (US EST)')
ax.set_ylabel('Porcentaje de envios para primera plana')
ax.set_title('Posibilidad de llegar a primera plana')
ax.grid('on')
```

```python
import numpy as np
import matplotlib.pyplot as plt 


x = np.arange(25)
y = np.random.randint(10, 20, 25)
y[24] = y[0]

fig, ax = plt.subplots(figsize=(7, 7))
ax.fill_between(x, y, color="skyblue", alpha=0.4, label="rea de posibilidad")
ax.plot(x, y, color="blue", marker="o", markersize=5, label="Tendencia")
ax.grid(True, linestyle='--', alpha=0.6)

# Etiquetas y ttulo
ax.set_xlabel('Hora de posteo (US EST)', fontsize=12)
ax.set_ylabel('Porcentaje de envos para primera plana', fontsize=12)
ax.set_title('Posibilidad de llegar a primera plana', fontsize=14, fontweight='bold')
ax.set_xlim(0, 24)
ax.set_ylim(8, 22)  
ax.legend(loc="upper right", fontsize=15)
plt.tight_layout()
plt.show()
```
