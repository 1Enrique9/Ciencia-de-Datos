Para elecciones grandes, agendas de medios utilizan encuestas de salida de votantes para ayudar a predecir el/la generadoro/a antes de que se cuenten los votos. Para las eleccion presidencial de 2020 de Estados Unidos, en un cuestionario de salida de votantes de California se obtuvo que el 65.5% contesto haber votado por Biden. La muestra fue solo de 2271 votantes de una poblacion adula de 17 M. Con esto se predijo que Biden ganaria California y obtendria sus votos del colegio electoral ¿Como esta encuesta de salida del 0.01% podria tener suficiente informacion para predecir el ganador? A tarves de simulacion ¿Como se podria decir si Baden gano el estado de California con base en esta muestra? 

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import norm

theta = 0.5  # Parámetro de la distribución Bernoulli
n = 2271        # Tamaño de la muestra
m = 10000     # Número de repeticiones




mu = theta
sigma_cuad = theta * (1 - theta)

medias_observadas = []


for i in range(m):
    # Generar una muestra aleatoria de tamaño n
    muestra = np.random.binomial(n=1, p=theta, size=n)
    media_observada = np.mean(muestra)
    medias_observadas.append(media_observada)

medias_observadas = np.array(medias_observadas)


sigma_cuad_media = sigma_cuad / n  # Varianza de la media muestral
sigma_media = np.sqrt(sigma_cuad_media)  # Desviación estándar de la media muestral


plt.hist(medias_observadas, bins=100, density=True, edgecolor='black', alpha=0.7, label="Medias observadas")


x = np.linspace(mu - 4*sigma_media, mu + 4*sigma_media, 1000)
pdf_teorica = norm.pdf(x, mu, sigma_media)

plt.plot(x, pdf_teorica, 'r-', label="PDF teórica (Normal)")
plt.title(f"Distribución de las medias observadas (n = {n}, θ = {theta})")
plt.xlabel("Media observada (x̄ₙ)")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```
